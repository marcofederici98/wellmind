```python
import pandas as pd
```

```python
df = pd.read_csv('../data/output/df_clean_final.csv')
```

```python
df = df.drop("Unnamed: 0", axis = 1)
```

```python
df
```

```python
repl_maybe = ["I don't know", 'Maybe', 'Maybe/Not sure', 'I am not sure']
maybe = 'Non lo so'
```

```python
for i in df:
    df[i] = df[i].replace(repl_maybe, maybe)
    df[i] = df[i].replace('Yes', 'Sì')
```

```python
df
```

```python
df['sex'] = df['sex'].replace('Male', 'Uomo')
df['sex'] = df['sex'].replace('Female', 'Donna')
```

```python
df['remote_flag'] = df['remote_flag'].replace('Sometimes', 'A volte')
df['remote_flag'] = df['remote_flag'].replace('Never', 'Mai')
df['remote_flag'] = df['remote_flag'].replace('Always', 'Sempre')
```

```python
df
```

```python
df.columns[11]
```

```python
df['prev_mh_benefits'].unique()
```

```python
df['prev_mh_benefits'] = df['prev_mh_benefits'].replace('No, none did', 'No, nessuno')
df['prev_mh_benefits'] = df['prev_mh_benefits'].replace('Some did', 'Alcuni')
df['prev_mh_benefits'] = df['prev_mh_benefits'].replace('Yes, they all did', 'Sì, tutti')
```

```python
df.to_csv('df_tradotto.csv')
```
