```python
diz_difficult = {'Very easy': 4,
 'Somewhat easy': 3,
 'Neither easy nor difficult': 2,
 'Very difficult': 0,
 'Somewhat difficult': 1}
diz_sharing = {'Somewhat open': 3,
 'Neutral': 2,
 'Very open': 4,
 'Not open at all': 0,
 'Somewhat not open': 1}
diz_repl = {'Sì': 1, 'No': -1, 'Non lo so': 0}
diz_remote = {'A volte': 1, 'Mai': 0, 'Sempre': 2}
diz_age = {'18-24': 0, '25-35': 1, '36-49': 2, '50+': 3}
diz_prev_ben = {'No, nessuno': 0, 'Sì, tutti': 3, 'Alcuni': 2, 'Non lo so': 1}
```

```python
l_diz = [diz_difficult, diz_repl, diz_remote, diz_age, diz_sharing]
```

```python
sex_m = []
sex_f = []
sex_b = []
for i in df['sex']:
    if i == 'Uomo':
        sex_m.append(1)
        sex_f.append(0)
        sex_b.append(0)
    elif i == 'Donna':
        sex_m.append(0)
        sex_f.append(1)
        sex_b.append(0)
    else:
        sex_m.append(0)
        sex_f.append(0)
        sex_b.append(1)
        
```

```python
def custom_enc(df):
    for i in df:
        if i == 'prev_mh_benefits':
            df[i] = df[i].replace(diz_prev_ben)
        for j in l_diz:
            df[i] = df[i].replace(j)
    df['sex_m'] = sex_m
    df['sex_f'] = sex_f
    df['sex_b'] = sex_b
    df = df.drop('sex', axis = 1)
    return df
```

```python
enc_df = custom_enc(df)
```

```python
diz_list = l_diz.copy()
```

```python
diz_list.append(diz_prev_ben)
```

```python
diz_inv = []
for i in diz_list:
    invertito = {}
    for chiave, valore in i.items():
        invertito[valore] = chiave
    diz_inv.append(invertito)
```

```python
diz_inv
```

```python
def rev_enc(df):
    for i in df:
        for j in diz_inv:
            df[i] = df[i].replace(j)
    return df
```
