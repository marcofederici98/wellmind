```python
import pandas as pd
import plotly.graph_objects as go
import plotly.express as px
import matplotlib.pyplot as plt
import mpld3
```

```python
df = pd.read_csv('../../data/output/df_tradotto.csv').drop('Unnamed: 0', axis = 1)
```

```python
def histplot(x, color = None, pattern_shape = None):
    fig = px.histogram(df, x = x, color = color, pattern_shape = pattern_shape)
    return fig.show()
```

```python
df.columns[0]
```

```python
for i in df.columns:
    print(histplot(i))
```

```python
x = df['mh_coverage_flag']
fig, ax = plt.subplots()
ax.pie(x= 'mh_coverage_flag', data = df)
plt.show()
```
